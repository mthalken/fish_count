```python
# Import the dependencies.
import pandas as pd
import matplotlib.pyplot as plt
from pathlib import Path
```

```python
file_path = Path('Resources/ColumbiaBasin_combined_count.csv')
df = pd.read_csv(file_path)
```

```python
df.head(12)
```

```python
hyphens = ["-"]
```

```python
for word in hyphens:
    df["Jan"] = df["Jan"].str.replace(word,"0")
    df["Feb"] = df["Feb"].str.replace(word,"0")
    df["Mar"] = df["Mar"].str.replace(word,"0")
    df["Apr"] = df["Apr"].str.replace(word,"0")
    df["May"] = df["May"].str.replace(word,"0")
    df["Jun"] = df["Jun"].str.replace(word,"0")
    df["Jul"] = df["Jul"].str.replace(word,"0")
    df["Aug"] = df["Aug"].str.replace(word,"0")
    df["Sep"] = df["Sep"].str.replace(word,"0")
    df["Oct"] = df["Oct"].str.replace(word,"0")
    df["Nov"] = df["Nov"].str.replace(word,"0")
    df["Dec"] = df["Dec"].str.replace(word,"0")
    df["Total"] = df["Total"].str.replace(word,"0")
```
