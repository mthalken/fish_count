```python
# Import the dependencies.
import pandas as pd
import matplotlib.pyplot as plt
from pathlib import Path
```

```python
file_path = Path('../Resources/refactored_data/ColumbiaBasin_combined_count.csv')
df = pd.read_csv(file_path, skipfooter=1, thousands=',')
```

```python
df
```

```python
for column in df.columns:
    if column != 'Location Name':
        hyphen_idx = df[column] == '-'
        df.loc[hyphen_idx, column] = 0
        df[column] = df[column].astype('int64')
```

```python
annual_count_df = df.groupby(['Year']).sum().reset_index()
annual_count_df.drop('Code', axis=1, inplace=True)
annual_count_df
```
