```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Create path and df
df = pd.read_csv('../Resources/total_data.csv')
df
```

```python
# Create df with specific columns
df_cut_week = df[['Week_Number', 'Stlhd', 'TempC']]
df_cut_week
```

```python
# create groupby for Stlhd by week number
week_df = df_cut_week.groupby(['Week_Number']).mean().reset_index()
week_df
```

```python
week_df.plot.bar('Week_Number', 'Stlhd')
```

```python
plt.plot(week_df.Week_Number, week_df.Stlhd)
plt.plot(week_df.Week_Number, week_df.TempC)
plt.show()
```

```python
# Create df with specific columns
df_cut_month = df[['Month', 'Stlhd', 'TempC']]
df_cut_month
```

```python
# create groupby for Stlhd by month
month_df = df_cut_month.groupby(['Month']).mean().reset_index()
month_df
```

```python
month_df.plot.bar('Month', 'Stlhd')
```

```python
plt.plot(month_df.Month, month_df.Stlhd)
plt.plot(month_df.Month, month_df.TempC)
plt.show()
```

```python
week_df.to_csv('../Resources/week_averages.csv')
```

```python
month_df.to_csv('../Resources/month_averages.csv')
```
