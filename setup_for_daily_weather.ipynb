```python
# Import dependencies
import pandas as pd
from pathlib import Path
import matplotlib.pyplot as plt
```

```python
# Load data into dataframe
file_path = Path('Resources/combineddata.csv')
df = pd.read_csv(file_path)
df.head(3)
```

```python
# Pull columns for weather data
df_weather = df[['daterecorded','maxtempf','mintempf','precipitationinch','watertempf']]
df_weather.head(3)
```

```python
df_weather.dtypes
```

```python
# Removing the year from 'daterecorded'
df_weather['daterecorded'] = df_weather['daterecorded'].str[5:]
```

```python
df_weather.head()
```

```python
df_weather.tail()
```

```python
# Group by 'Date' to get 30 years of average weather on the day
# Anticipate 366 row due to leap years
daily_ave_df = df_weather.groupby(['daterecorded']).mean().reset_index()
daily_ave_df
```

```python
# Write to .csv
daily_ave_df.to_csv('Resources/dailyWeatherTable.csv', index=False)
```
